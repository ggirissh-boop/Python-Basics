Question 1 : What is Python and why is it widely used in Data Analytics? Mention any three reasons.

Answer :- Python is a high‑level, open‑source programming language known for its simplicity, versatility, and powerful ecosystem of data‑focused libraries. It is widely used in Data Analytics because it makes data cleaning, analysis, visualization, and machine‑learning workflows faster and easier.
---
Why Python is widely used in Data Analytics
Below are three strong, source‑supported reasons:
1. Easy to learn and read — Python has a simple, intuitive syntax, making it beginner‑friendly and ideal for analysts transitioning from Excel or SQL.
2. Rich ecosystem of data libraries — Libraries like Pandas, NumPy, Matplotlib, SciPy, and Scikit‑learn make data manipulation, statistical analysis, visualization, and machine learning efficient and standardized.
3. Large and active community support — Python has one of the strongest global communities, continuously contributing tools, documentation, and updates that help analysts solve problems quickly.

Question 2 : Explain the difference between List and Tuple in Python.

Answer :- A List in Python is mutable, uses square brackets [], and is suitable for data that needs frequent modification. A Tuple is immutable, uses parentheses (), and is preferred for fixed data and faster performance. Tuples also consume less memory and have fewer built‑in methods.

Question 3 : What is a function in Python? Why are functions useful?

Answer :- A function is a named block of code that runs only when it is called.
It can take inputs (parameters), perform operations, and optionally return a value.
1. Reusability
You write code once and use it multiple times, reducing duplication.
2. Improves Code Organization
Functions break a large program into smaller, manageable parts—making it easier to read and maintain.
3. Enhances Debugging and Testing
Since each function handles a specific task, errors are easier to locate and fix.

Question 5 : Write a Python program to check whether a number is even or odd using conditional statements.

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print(num, "is Even")
else:
    print(num, "is Odd")
```

Question 8 : Write a Python program to remove duplicate values from a list using a set.

```python
numbers = [10, 20, 10, 30, 20, 40]

unique_numbers = list(set(numbers))

print("List after removing duplicates:", unique_numbers)
```

Question 9 : Write a function that returns the square of a number.

```python
number = int(input("Enter a number: "))

def square(number):
    return number ** 2

print("The square of", number, "is", square(number))
```
